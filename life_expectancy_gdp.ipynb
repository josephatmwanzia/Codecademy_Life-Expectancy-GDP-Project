```python
#import dependancies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load the data to a dataframe and inspect it.
gdp_data = pd.read_csv(r"C:\Users\jmwanzia001\Documents\GitHub\Codecademy_Life-Expectancy-GDP-Project\all_data.csv")
gdp_data.columns = ['country', 'year', 'life_exp_years', 'GDP']
print(gdp_data.head())
print(gdp_data.country.unique())
```

```python
#Explore life expectancy
f, ax = plt.subplots()
ax = sns.lineplot(data= gdp_data, x = 'year', y= 'life_exp_years', hue= 'country')
ax.set_title('Life Expectancy for 6 countries')
plt.show()

#Global life expectancy has generaly increased between 2000 and 2015.
#Zimbabwe has comparatively lower GDP. This has however improved significantly.
```

```python
#Explore GDP
ax2 = plt.subplots()
ax2 = sns.lineplot(data= gdp_data, x = 'year', y= 'GDP', hue= 'country')
ax2.set_title('GDP for 6 countries')
plt.show()

#USA has significantly biger GDP compared to the other countries.
#Chinas GDP has grown significantly in the 15 years.
```

```python
#Explore GDP without china and USA
ax3 = plt.subplots()
ax3 = sns.lineplot(data= gdp_data[(gdp_data.country != 'China') & (gdp_data.country != 'United States of America')],
                   x = 'year', y= 'GDP', hue= 'country')
ax3.set_title('GDP for 4 countries')
plt.show()

#Germany GDP grew between 2002 and 2008 but then stagnated.
#Mexico GDP has had a small but systained growth.
#Chile and Zimbabwe GDPs have remained relatively flat.
```
